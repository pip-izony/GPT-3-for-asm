```
from api import GPT,Example
gpt = GPT(temperature=0.1, max_tokens=15000)

gpt.add_example(Example(
    "function_1",
    "section .data\n\
msg db \"Hello_World\", 0x0A\n\
section .text\n\
global _start\n\
_start:\n\
mov rax, 1\n\
mov rdi, 1\n\
mov rsi, msg\n\
mov rdx, 11\n\
je _exit\n\
syscall\n\
_exit:\n\
mov rax, 60\n\
mov rdi, 0\n\
syscall"))

gpt.add_example(Example(
    "function_2",
    "section .data\n\
	star: db '*********'\n\
	new_line_value: db 0x0A\n\
section .text\n\
global _start\n\
_start:\n\
	push rbp\n\
	mov rbp, rsp\n\
	mov r10, 4\n\
	mov rdx, 1\n\
	cmp r10, 0\n\
	je _zero_loop\n\
	mov r8, 1\n\
	mov r9, 1\n\
_loop:\n\
	cmp r9, 0\n\
	je _exit\n\
	cmp r9, r10\n\
	je _change_inc_to_dec\n\
_print_all:\n\
	mov rdx, r9\n\
	call _star_print\n\
	call _new_line\n\
	add r9, r8\n\
	call _loop\n\
_change_inc_to_dec:\n\
	mov r8, -1\n\
	call _print_all\n\
_new_line:\n\
	mov rax, 1\n\
	mov rdi, 1\n\
	mov rsi, new_line_value\n\
	mov edx, 1\n\
	syscall\n\
	ret\n\
_star_print:\n\
	mov rax, 1\n\
	mov rdi, 1\n\
	mov rsi, star\n\
	syscall\n\
	ret\n\
_zero_loop:\n\
	mov rax, 1\n\
	mov rdi, 1\n\
	mov rsi, star\n\
	syscall\n\
	call _new_line\n\
	call _exit\n\
_exit:\n\
	mov rax, 60\n\
	syscall"))

gpt.add_example(Example(
    "function_1 + function_2",
    "section .data\n\
msg db \"Hello_World\", 0x0A\n\
section .text\n\
global _start\n\
_start:\n\
mov rax, 1\n\
mov rdi, 1\n\
mov rsi, msg\n\
mov rdx, 11\n\
je _exit\n\
syscall\n\
_exit:\n\
mov rax, 60\n\
mov rdi, 0\n\
syscall\n\
section .data\n\
	star: db '*********'\n\
	new_line_value: db 0x0A\n\
section .text\n\
global _start\n\
_start:\n\
	push rbp\n\
	mov rbp, rsp\n\
	mov r10, 4\n\
	mov rdx, 1\n\
	cmp r10, 0\n\
	je _zero_loop\n\
	mov r8, 1\n\
	mov r9, 1\n\
_loop:\n\
	cmp r9, 0\n\
	je _exit\n\
	cmp r9, r10\n\
	je _change_inc_to_dec\n\
_print_all:\n\
	mov rdx, r9\n\
	call _star_print\n\
	call _new_line\n\
	add r9, r8\n\
	call _loop\n\
_change_inc_to_dec:\n\
	mov r8, -1\n\
	call _print_all\n\
_new_line:\n\
	mov rax, 1\n\
	mov rdi, 1\n\
	mov rsi, new_line_value\n\
	mov edx, 1\n\
	syscall\n\
	ret\n\
_star_print:\n\
	mov rax, 1\n\
	mov rdi, 1\n\
	mov rsi, star\n\
	syscall\n\
	ret\n\
_zero_loop:\n\
	mov rax, 1\n\
	mov rdi, 1\n\
	mov rsi, star\n\
	syscall\n\
	call _new_line\n\
	call _exit\n\
_exit:\n\
	mov rax, 60\n\
	syscall"))


prompt = "function_3 + function_4"
output = gpt.submit_request(prompt)
output.choices[0]['text']
```
